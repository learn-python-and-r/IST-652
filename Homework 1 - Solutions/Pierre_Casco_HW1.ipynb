Data Question:
Does family income predict how much of a donation is made?

The plot shows the average home value on the x-axis and average family income on the y-axis. The point color represent the average gift made. The highest average donations are plotted near the y-axis. People with les family income tend to donate more money per average.

```python
px.scatter(df,x = 'Avg_Home_Value',y = 'Average_Family_Income',color = 'Avg_Gift')
```

In order to see the percentages of donations, the donation amount was divided by the average family income. These percentages were plotted on a bar chart to see where the largest percentages of donations come from.

```python
# Create column showing percent of donation made
df['Percent_Donation'] = (df['Donation_Amt']/df['Median_Family_Income'])*100
```

The bar chart shows that the highest percentage of donations were between 200 and 600.

```python
px.bar(df,x='Median_Family_Income',y='Donation_Amt')
```

The next data question is whether the amount of children makes a difference in donations. To view this, a scatter plot was created with donations on the y-axis and number of children on the x-axis. The plot showed the highest donation amounts came from families with only one child. THe least amount of donations came from the families with five kids. 

```python
px.scatter(df,x='Number_of_Children', y = 'Donation_Amt')
```

For the output file, all the people with 1 child who have donated previously were printed because they donate the most amount of money. The file was written using the pandas library method "to_csv".

```python
# Create new dataframe with families with 1 child who have made a donation in the past
one_child = df[(df['Number_of_Children'] == 1) & (df['Donor'] == 1)]
```

```python
# Write file to csv
one_child.to_csv('Donors with 1 Child.csv')
```
